```python
Analizamos si por lote existe alguna relación entre:


coste_inicial (lo que cuesta producir o comprar)

precio_venta (a cuánto se vende)

rentabilidad (en %)
```

```python
df_tipo_limpio = df_tipo.dropna(subset=['coste_inicial']).copy()
df_tipo_limpio = df_tipo_limpio[df_tipo_limpio['coste_inicial'] != 0]
df_tipo_limpio['rentabilidad'] = ((df_tipo_limpio['precio_venta'] - df_tipo_limpio['coste_inicial']) / df_tipo_limpio['coste_inicial']) * 100

df_lote = df_tipo_limpio.groupby('lote', as_index=False)[['coste_inicial', 'precio_venta', 'rentabilidad']].mean()
corr_lote = df_lote[['coste_inicial', 'precio_venta', 'rentabilidad']].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
sns.heatmap(corr_lote, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title('Correlaciones Económicas por Lote')
plt.tight_layout()
plt.show()
```

Los lotes con costes altos también tienen precios altos, pero cuando aumenta el coste, la rentabilidad tiende a disminuir.
El precio de venta no influye en la rentabilidad.(?)



```python
tipos = df_tipo['tipo'].unique()
resumen = []

for fruta in tipos:
    sub = df_tipo[df_tipo['tipo'] == fruta].copy()
    sub = sub.dropna(subset=['coste_inicial', 'precio_venta'])
    sub = sub[sub['coste_inicial'] != 0]
    sub['rentabilidad'] = ((sub['precio_venta'] - sub['coste_inicial']) / sub['coste_inicial']) * 100

    df_lote = sub.groupby('lote', as_index=False)[['coste_inicial', 'precio_venta', 'rentabilidad']].mean()
    corr = df_lote[['coste_inicial', 'precio_venta', 'rentabilidad']].corr()

    resumen.append({
        'tipo': fruta,
        'corr_coste_precio': corr.loc['coste_inicial', 'precio_venta'],
        'corr_coste_rentabilidad': corr.loc['coste_inicial', 'rentabilidad'],
        'corr_precio_rentabilidad': corr.loc['precio_venta', 'rentabilidad'],
        'rentabilidad_media': sub['rentabilidad'].mean()
    })

    plt.figure(figsize=(8,5))
    sns.histplot(sub['rentabilidad'], bins=30, kde=True)
    plt.title(f'Distribución de Rentabilidad — {fruta}')
    plt.xlabel('Rentabilidad (%)')
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(5,4))
    sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt=".2f")
    plt.title(f'Correlaciones Económicas por Lote — {fruta}')
    plt.tight_layout()
    plt.show()

df_resumen = pd.DataFrame(resumen)
display(df_resumen.sort_values('rentabilidad_media', ascending=False))
```
